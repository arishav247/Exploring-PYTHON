```python
""" Pandas: Introduction
The pandas is an open source data analysis and manipulation tool,
built on top of the Python programming language.
Used in data pre-processing, plotting, descriptive statistics  
Two basic data structures: Series and Data Frame
"""
```

```python
import pandas as pd

listSeries = [1, 2, -3, 4.5, "Python"]
print(listSeries)

##[1, 2, -3, 4.5, 'Python']
##dtype: object

seriesPanda1 = pd.Series(listSeries)
print(seriesPanda1)
print(type(seriesPanda1))
```

```python
seriesPanda2 = pd.Series([1,2,3,5])
print(seriesPanda2)

""" Series with customized indexing"""
seriesPanda3 = pd.Series([1,2,3,-5.5], index = ['a','b','c','d'])
print(seriesPanda3)

""" Mentioned the data type and the series name"""
seriesPanda4 = pd.Series([1,2,3,5], dtype = float, name = "Series Panads Data")
print(seriesPanda4)


seriesPanda5 = pd.Series(8, index = [1,2,3])
print(seriesPanda5)
```

```python
""" Dictionary use"""
seriesPanda6 = pd.Series({'x':10, 'y':11,})
print(seriesPanda6)

seriesPanda7 = pd.Series([10,20,30,40])
seriesPanda8 = pd.Series([1,2,3,4])
seriesPanda9 = pd.Series([1,2])

""" Print the value of an index"""
print("The value of 1st index: ",seriesPanda7[1])
##The value of 1st index:  2

""" Addition of two series"""
print(seriesPanda7 + seriesPanda8)
```

```python
""" To get maximum and minimum """
print("Minimum, Maximum & Sum: ", min(seriesPanda7),max(seriesPanda7),sum(seriesPanda7))
#Minimum, Maximum & Sum:  10 40 100

""" Print on conditions"""
print(seriesPanda7[seriesPanda7<30])
```

## Data Frame

```python
"""##   Data Frame ###"""
""" Pandas: Introduction (Part 2)
The pandas is an open source data analysis and manipulation tool,
built on top of the Python programming language.
Used in data pre-processing, plotting, descriptive statistics  
Two basic data structures: Series and Data Frame

pd.DataFrame(data, index (Optional), columns (Optional), dtype (Optional), copy(optional))
"""
""" Create an empty data frame"""
```

```python
dfPanda1 = pd.DataFrame()
print(dfPanda1) # Empty data frame

dfPanda2 = pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]])
print(dfPanda2)

dfPanda3 = pd.DataFrame({'Sr':[1,2,3,4]})
print(dfPanda3)

dfPanda4 = pd.DataFrame({'ID':['a','b','c','d'], 'Sr':[1,2,3,4]})
print(dfPanda4)

dfPanda5 = pd.DataFrame([{'x':1,'y':2},{'x':11,'y':12},{'x':22,'y':23}])
print(dfPanda5)

dfPanda6 = pd.DataFrame({'A':pd.Series([10,11,12]),'B':pd.Series(['x','y','z'])})
print(dfPanda6)
```
